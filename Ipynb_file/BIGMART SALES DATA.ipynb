```python
import pandas as pd
```

```python
# Load the data set
COG = pd.read_csv("Test.csv")
```

```python
# View the dataframe
COG
```

```python
# View the columns
COG.columns
```

```python
# Check for duplicates
COG.duplicated().sum()
```

```python
# Finding empty rows
COG.isna().sum()
```

```python
# Treating our missing values accordingly
mean = COG["Item_Weight"].mean()
COG["Item_Weight"].fillna(mean, inplace=True)
mode_outlet_size = COG["Outlet_Size"].mode()[0]
COG["Outlet_Size"].fillna(mode_outlet_size, inplace=True)
COG.isna().sum()
```

```python
# View the summary statistics of the dataset
COG.describe()
```

```python
# View the general info of the tableZ
COG.info()
```

```python
# Changing our int values to 2 decimal places
COG['Item_MRP'] = COG['Item_MRP'].round(2)
COG['Item_Weight'] = COG['Item_Weight'].round(2)
COG['Item_Visibility'] = COG['Item_Visibility'].round(2)
```

```python
COG
```

```python
average_mrp = COG.groupby('Item_Type')['Item_MRP'].mean()
average_visibility = COG.groupby('Item_Type')['Item_Visibility'].mean()
average_mrp
```

```python
average_visibility
```

#### Average MRP And Visibility By Item Type

```python
# Creating a figure with 1 row and 2 columns for side-by-side plots
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plotting the Average MRP by Item Type on the first subplot
average_mrp.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_title('Average MRP by Item Type')
ax1.set_xlabel('Item Type')
ax1.set_ylabel('Average MRP')
ax1.tick_params(axis='x', rotation=110)

# Plotting the Average Visibility by Item Type on the second subplot
average_visibility.plot(kind='bar', color='coral', ax=ax2)
ax2.set_title('Average Visibility by Item Type')
ax2.set_xlabel('Item Type')
ax2.set_ylabel('Average Visibility')
ax2.tick_params(axis='x', rotation=110)

# Adjusting layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Calculate average Item_MRP and Item_Visibility by Outlet_Size
avg_mrp_visibility = COG.groupby('Outlet_Size')[['Item_MRP', 'Item_Visibility']].mean().reset_index()
# Bar plot for average MRP by Outlet Size
import seaborn as sns
plt.figure(figsize=(10, 5))
A = sns.barplot(x='Outlet_Size', y='Item_MRP', data=avg_mrp_visibility, palette='viridis')
plt.title('Average MRP by Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Average MRP')
# Add values on top of each bar
for index, row in avg_mrp_visibility.iterrows():
    A.text(index, row['Item_MRP'] + 1, round(row['Item_MRP'], 1), color='black', ha="center")

plt.show()

# Box plot for visibility distribution by Outlet Size
plt.figure(figsize=(10, 5))
sns.boxplot(x='Outlet_Size', y='Item_Visibility', data=COG, palette='viridis')
plt.title('Item Visibility by Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Item Visibility')
plt.show()
```

```python
# Calculate the count of unique Item_Types for each Outlet_Establishment_Year
product_variety = COG.groupby('Outlet_Establishment_Year')['Item_Type'].nunique().reset_index()

# Scatter plot for Outlet Age and Product Variety
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Outlet_Establishment_Year', y='Item_Type', data=product_variety, color='b')
plt.title('Outlet Age and Product Variety')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Number of Unique Product Types')
plt.show()
```

```python
# Calculate the count of each Item_Type for each Outlet_Location_Type
product_distribution = COG.groupby(['Outlet_Location_Type', 'Item_Type']).size().unstack()

# Stacked bar chart
product_distribution.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="viridis")
plt.title('Distribution of Product Types by Location Type')
plt.xlabel('Outlet Location Type')
plt.ylabel('Number of Items')
plt.legend(title='Item Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Assuming your data has columns 'Outlet_Type' and 'Item_MRP'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Outlet_Type', y='Item_MRP', data=COG)
plt.title('MRP Distribution Across Outlets')
plt.xlabel('Outlet Type')
plt.ylabel('Item MRP')
plt.xticks(rotation=120)  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_mrp_by_outlet = COG.groupby('Outlet_Type')['Item_MRP'].mean().reset_index()
A = sns.barplot(x='Outlet_Type', y='Item_MRP', data=avg_mrp_by_outlet, palette='viridis')
plt.title('Average MRP by Outlet Type')
plt.xlabel('Outlet Type')
plt.ylabel('Average MRP')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
# Add values on top of each bar
for index, row in avg_mrp_visibility.iterrows():
    A.text(index, row['Item_MRP'] + 1, round(row['Item_MRP'], 1), color='black', ha="center")

plt.show()
```

```python
plt.figure(figsize=(12, 6))
avg_mrp_by_year = COG.groupby('Outlet_Establishment_Year')['Item_MRP'].mean().reset_index()
sns.lineplot(x='Outlet_Establishment_Year', y='Item_MRP', data=avg_mrp_by_year)
plt.title('Average MRP Trends by Establishment Year')
plt.xlabel('Outlet_Establishment_Year')
plt.ylabel('Average MRP')
plt.show()
```

```python
# Group data by year and item type to see trends in the count of each item type over time
item_type_trends = COG.groupby(['Outlet_Establishment_Year', 'Item_Type']).size().unstack().fillna(0)

# Plot changes in product mix by year
plt.figure(figsize=(14, 8))
item_type_trends.plot(kind='line', marker='o', ax=plt.gca())
plt.title("Changes in Product Mix by Year")
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Count of Items")
plt.legend(title='Item Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Group data by year and fat content to see trends in fat content distribution over time
fat_content_trends = COG.groupby(['Outlet_Establishment_Year', 'Item_Fat_Content']).size().unstack().fillna(0)

# Plot changes in fat content distribution by year
plt.figure(figsize=(10, 6))
fat_content_trends.plot(kind='line', marker='o', ax=plt.gca())
plt.title("Changes in Fat Content Distribution by Year")
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Count of Items")
plt.legend(title='Fat Content', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
